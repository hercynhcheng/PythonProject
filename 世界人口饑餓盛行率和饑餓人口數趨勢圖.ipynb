```python
import matplotlib.pyplot as plt
# :https://www.fao.org/hunger/en/
#:
# 
x = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
y2 = [586.2, 584.7, 570.7, 571.6, 588.6, 585.1, 573.3, 590.6, 618.4, 721.7, 767.9]
y1 = [8.3, 8.2, 7.9, 7.8, 8, 7.8, 7.6, 7.7, 8.2, 8, 9.3]

# 
fig, ax1 = plt.subplots()

# 
ax1.plot(x, y1, marker='^', color='red')  #  marker 
ax1.set_xlabel('year')
ax1.set_ylabel('Prevalence of undernourishment (percentage)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# 
ax2 = ax1.twinx()
ax2.plot(x[:len(y2)], y2, marker='x', color='blue')  #  marker 
ax2.set_ylabel('Number of undernourished (millions)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# 
plt.title('Trend Chart of Prevalence of Undernourishment and Number of Undernourished Population')

# 
plt.show()
```
